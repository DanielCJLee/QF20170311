```python
%matplotlib inline
import random
import matplotlib.pyplot as plt

duration = 100
mean = 0.0
std = 1.0

x = range(duration)
y = []
yNow = 0

for i in x:
    nextDelta = random.normalvariate(mean, std)
    yNow += nextDelta
    y.append(yNow)

plt.plot(x,y)
plt.xlabel("Time")
plt.ylabel("Value")
plt.savefig("RandomGraph.png")
plt.show()
```
